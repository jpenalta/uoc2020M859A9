```python
#pd.set_option('display.max_columns', None)
#df_schools.tail()

df = df_schools[['CNTRYID','SC001Q01TA', 'SC013Q01TA', 'SC016Q01TA', 'SC016Q02TA','SC016Q03TA','SC016Q04TA']].copy()

df['SC016Q01TA'] = pd.to_numeric(df['SC016Q01TA']).fillna(0)
df['SC016Q02TA'] = pd.to_numeric(df['SC016Q02TA']).fillna(0)
df['SC016Q03TA'] = pd.to_numeric(df['SC016Q03TA']).fillna(0)
df['SC016Q04TA'] = pd.to_numeric(df['SC016Q04TA']).fillna(0)



one_hot = pd.get_dummies(df[['SC001Q01TA', 'SC013Q01TA']])

df.drop(['SC001Q01TA'],axis=1, inplace=True)
df.drop(['SC013Q01TA'], axis=1, inplace=True)
df = df.join(one_hot)

columns= ['CNTRYID', 
        'Gobierno', 'Padres', 'Benefactores', 'Otros',
        'SC001Q01TA_Ciudad entre 100.000 y 1.000.000 habitantes',
        'SC001Q01TA_Gran ciudad con ms de 1.000.000 habitantes',
        'SC001Q01TA_Pequeo pueblo, entre 3.000 y 15 000 habitantes',
        'SC001Q01TA_Pueblo entre 15.000 y 100.000 habitantes',
        'SC001Q01TA_Villa o rea rural, menos de 3.000 habitantes',
        'Escuela privada',
        'Escuela pblica']
df.columns = columns

df.head()


#df[['SC001Q01TA']].tail()

```

# Financiacin

```python
df_agrupado = df.groupby(['CNTRYID']).sum()

total_finaciacion = df_agrupado[['Gobierno', 'Padres', 'Benefactores', 'Otros']].sum(axis=1)

df_agrupado['Gobierno'] = df_agrupado['Gobierno']/total_finaciacion *100
df_agrupado['Padres'] = df_agrupado['Padres']/total_finaciacion *100
df_agrupado['Benefactores'] = df_agrupado['Benefactores']/total_finaciacion *100
df_agrupado['Otros'] = df_agrupado['Otros']/total_finaciacion * 100

total_ciudades = df_agrupado[columns[4:10]].sum(axis=1)

df_agrupado[columns[5]] = df_agrupado[columns[5]]/total_ciudades *100
df_agrupado[columns[6]] = df_agrupado[columns[6]]/total_ciudades *100
df_agrupado[columns[7]] = df_agrupado[columns[7]]/total_ciudades *100
df_agrupado[columns[8]] = df_agrupado[columns[8]]/total_ciudades *100
df_agrupado[columns[9]] = df_agrupado[columns[9]]/total_ciudades *100

total_tipo = df_agrupado[columns[10:11]].sum(axis=1)
df_agrupado[columns[10]] = df_agrupado[columns[10]]/total_ciudades *100
df_agrupado[columns[11]] = df_agrupado[columns[11]]/total_ciudades *100


df_agrupado.head()
```

```python
df_agrupado.to_csv('./dato/financiacion.csv')
```

# Recursos TIC

Extraccin de los datos de los recursos TIC disponibles para los alumnos.

```python
df = df_schools[['CNTRYID','code','SC018Q01TA01','SC018Q01TA02','SC004Q01TA','SC004Q02TA'
                 ,'SC004Q03TA','SC004Q04NA','SC004Q05NA','SC004Q06NA','SC004Q07NA']].copy()
df.dropna(subset = ["SC004Q01TA"], inplace=True)


df['SC018Q01TA01'] = pd.to_numeric(df['SC018Q01TA01']).fillna(0)
df['SC018Q01TA02'] = pd.to_numeric(df['SC018Q01TA02']).fillna(0)
df['SC018Q01TA01'] = df['SC018Q01TA01'] + df['SC018Q01TA02']
df.drop('SC018Q01TA02', inplace=True, axis=1)


df['SC004Q01TA'] = (pd.to_numeric(df['SC004Q01TA']).fillna(0))
df['SC004Q02TA'] = pd.to_numeric(df['SC004Q02TA']).fillna(0)
df['SC004Q03TA'] = pd.to_numeric(df['SC004Q03TA']).fillna(0)
df['SC004Q04NA'] = pd.to_numeric(df['SC004Q04NA']).fillna(0)
df['SC004Q05NA'] = pd.to_numeric(df['SC004Q05NA']).fillna(0)
df['SC004Q06NA'] = pd.to_numeric(df['SC004Q06NA']).fillna(0)
df['SC004Q07NA'] = pd.to_numeric(df['SC004Q07NA']).fillna(0)


df_groupby = df.groupby(['CNTRYID'], as_index=False).mean()

df_groupby.insert(1,'code','')
for index, pais in df_paises.iterrows():
    df_groupby.loc[df_groupby['CNTRYID'] == pais['name'], ['code']] = pais['code']


columns= ['name', 
          'code',
          'Profesorado',
         'Alumnado',
         'Ordenadores',
         'Con conexin a internet',
         'Portatiles',
          'Pizarras interactivas',
          'Proyectores',
          'Ordenadores para profesores'
         ]


df_groupby.columns= columns

for col in columns:
    if col not in ['Alumnado','name','code','Profesorado','Ordenadores para profesores']:
       df_groupby[col] = df_groupby[col]/df_groupby['Alumnado']

df_groupby['Ordenadores para profesores'] = df_groupby['Ordenadores para profesores']/df_groupby['Profesorado']


df_groupby.head()

df_groupby.to_csv('./docs/data/recursostic.csv',, index=False)
```

# Capacidades TIC

```python
df = df_schools[['CNTRYID','code','SC155Q01HA','SC155Q02HA','SC155Q03HA','SC155Q04HA'
                 ,'SC155Q05HA','SC155Q06HA','SC155Q07HA','SC155Q08HA','SC155Q09HA'
                 ,'SC155Q10HA', 'SC155Q11HA']].copy()


#df['code'] = df['SC155Q08HA'].astype('category').cat.codes


for key in df.columns:
    if key != 'CNTRYID' and key !='code':
        values = ['Strongly disagree', 'Disagree','Agree','Strongly agree']
        for i,value in enumerate(values):
            df.loc[ df[key]==value, [key+"_code"]] = i
    
df.head()

df_groupby = df.groupby(['CNTRYID'],  as_index=False).mean()

df_groupby.insert(1,'code','')
for index, pais in df_paises.iterrows():
    df_groupby.loc[df_groupby['CNTRYID'] == pais['name'], ['code']] = pais['code']

columns= ['name', 'code',
         'Dispositivos conectados a internet',
          'Ancho de banda',
          'Dispositivos para docencia',
          'Capacidad de los dispositivos',
          'Software adecuado',
          'Tcnica y capacidades pedaggicas del profesorado',
          'Tiempo para preparar integrar TIC',
          'Recursos para aprender a usar dispositivos TIC',
          'Disponible plataforma de elearning',
          'Incentivos para integrar  dispositivos TIC ',
          'Personal tcnico cualificado']
df_groupby.columns = columns    

    
df_groupby.head()

df_groupby.to_csv('./pagina2/data/capacidadestic.csv', index=False)
```
